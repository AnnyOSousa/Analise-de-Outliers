```python
# Chamando as bibliotecas
import pandas as pd
import numpy as np
```

```python
# Puxando dataset de casas para alugar
houses_to_rent = pd.read_csv('/content/houses_to_rent.csv')
```

```python
# Visualizando as primeiras linhas do dataset
houses_to_rent.head()
```

```python
# Visualizando as estatísticas
houses_to_rent.describe()
```

```python
# Verificando na coluna banheiros valores discrepantes
houses_to_rent['bathroom'].value_counts()
```

```python
# Verificando IQR (Distancia inter quartil)
Q1 = np.quantile(houses_to_rent['bathroom'], 0.25)
Q3 = np.quantile(houses_to_rent['bathroom'], 0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Filtrando apenas os outliers
houses_to_rent[(houses_to_rent['bathroom'] < (Q1 - 1.5 * IQR)) | (houses_to_rent['bathroom'] > (Q3 + 1.5 * IQR))]
```

```python
# Verificando os inliers e visualizando
houses_to_rent_inleiers = houses_to_rent[(houses_to_rent['bathroom'] >= (Q1 - 1.5 * IQR)) & (houses_to_rent['bathroom'] <= (Q3 + 1.5 * IQR))]
houses_to_rent_inleiers.head()
```

```python
# Visualizando as estatísticas do inleiers
houses_to_rent_inleiers.describe()
```

Com a retirada dos outliers a uma diferença significativa em dados como a média
